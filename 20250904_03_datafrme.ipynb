```python
import pandas as pd
```

```python
name = pd.Series(["小陈", "小李", "小王", "小张", "小赵", "小周"], index=["001", "002", "003", "004", "005", "006"])
gender = pd.Series(["女", "女", "男", "男", "女", "男"], index=["006", "005", "004", "003", "002", "001"])
height = pd.Series([172.5, 168.0, 178.2, 181.3, 161.7, 159.8], index=["001", "002", "003", "004", "005", "006"])
grade = pd.Series([89, 92, 82, 96, 93, 84], index=["001", "002", "003", "004", "005", "006"])
df1 = pd.DataFrame({"姓名": name, "性别": gender, "身高": height, "成绩": grade})
df1
```

#### 更新DataFrame的一列值

需要注意的是,新更改的数据需要有index才能和原有数据对齐

否则所有标签无法对齐的地方,都会产生缺失值

```python
df1["成绩"] = pd.Series([90, 91, 83, 95, 94, 85], 
                      index=["001", "002", "003", 
                             "004", "005", "006"])
df1
```

赋值成Python列表也是可以的,不需要标签,会按顺序对齐,行数吻合就可以

```python
df1['成绩'] = [90, 91, 83, 95, 94, 85]
df1
```

#### 给DataFrame增加一列值
如果我们赋值的时候列名不存在呢?这样会多出一列

```python
df1['班级'] = ['一班','三班','二班','三班','一班','二班']
df1
```

#### 更新DataFrame一行值

series:把series的标签与列名相对应,不然会产生缺失值

```python
df1.loc['005'] = pd.Series(['小赵','女',162.7,95,'一班'], index = ['姓名','性别','身高','成绩','班级'])
df1
```

赋值给列表会省事一些

```python
df1.loc['005'] = ['小赵','女',162.7,95,'一班']
df1
```

#### 添加DataFrame的一行值
添加行只能用loc,不可以用iloc,不然会报错
给iloc的位置索引得是存在的

```python
df1.loc['007'] = ['小孙','男',182.7,71,'一班']
df1
```

#### 删除DataFrame的行
用.drop()

```python
df1.drop('003')
```

删除多行,放入列表,里面放多个标签索引

```python
df1.drop(['003','007'])
```

这个方法只会返回删除后的DataFrame但不会更改原来的DataFrame

```python
df1
```

如果确实想要让DataFrame那一行被删除,可以把变量赋值给drop返回的结果

```python
df1 = df1.drop(["003", "007"])
df1
```

#### 删除DataFrame的列

.drop('列名', 可选参数axis = 1)

```python
df1.drop('身高', axis = 1)
```

axis = 0 表示沿着索引纵向操作

axis = 1 表示沿着列名横向操作

一般默认axis = 0,数据的变量含义和类型一般是相通的,比如都是描述身高的浮点数,所以适合进行统计操作

drop默认axis = 0,沿着索引纵向进行操作,删除的就是行

axis = 1时,是沿着列名横向操作,所以删除的就是列

#### 一次性删除多列

索引和索引,列名和列名也会去对齐

```python
df1.drop(['性别','身高'], axis = 1)
```

#### 和单个数字运算

```python
name = pd.Series(["小陈", "小李", "小王", "小张", "小赵", "小周"], index=["001", "002", "003", "004", "005", "006"])
gender = pd.Series(["女", "女", "男", "男", "女", "男"], index=["006", "005", "004", "003", "002", "001"])
height = pd.Series([172.5, 168.0, 178.2, 181.3, 161.7], index=["001", "002", "003", "004", "005"])
students = pd.DataFrame({"姓名": name, "性别": gender, "身高": height})
students
```

```python
students * 5
```

也可以提取出需要的DataFrame数据相乘

```python
students[['性别','身高']] * 5
```
